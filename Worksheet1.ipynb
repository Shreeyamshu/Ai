```python
#Array Creation

import numpy as np

# 1 Initialize an empty array with size 2x2
empty_array = np.empty((2, 2))
print("1. Empty array (2x2):")
print(empty_array)

# 2 Initialize an all one array with size 4x2
ones_array = np.ones((4, 2))
print("\n2. All ones array (4x2):")
print(ones_array)

# 3 Return new array filled with fill_value
filled_array = np.full((3, 3), 7)  # 3x3 array filled with 7
print("\n3. Array filled with value 7:")
print(filled_array)

# 4 Return zeros with same shape and type as given array
sample_array = np.array([[1, 2], [3, 4]], dtype=float)
zeros_like_array = np.zeros_like(sample_array)
print("\n4. Zeros like sample array:")
print(zeros_like_array)

# 5 Return ones with same shape and type as given array
ones_like_array = np.ones_like(sample_array)
print("\n5. Ones like sample array:")
print(ones_like_array)

# 6 Convert list to numpy array
new_list = [1, 2, 3, 4]
list_to_array = np.array(new_list)
print("\n6. List converted to array:")
print(list_to_array)
```

```python
#Array Manipulation: Numerical Ranges and Array indexing

import numpy as np

# 1 Array with values from 10 to 49
arr_10_49 = np.arange(10, 50)
print("1. Array from 10 to 49:")
print(arr_10_49)

# 2 3x3 matrix with values 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n2. 3x3 matrix (0-8):")
print(matrix_3x3)

# 3 3x3 identity matrix
identity_3x3 = np.eye(3)
print("\n3. 3x3 identity matrix:")
print(identity_3x3)

# 4 Random array of size 30 and mean
random_arr = np.random.random(30)
random_mean = random_arr.mean()
print(f"\n4. Random array mean: {random_mean:.4f}")

# 5 10x10 random array with min/max
random_10x10 = np.random.random((10, 10))
min_val = random_10x10.min()
max_val = random_10x10.max()
print(f"5. 10x10 array - Min: {min_val:.4f}, Max: {max_val:.4f}")

# 6 Zero array with 5th element replaced
zero_arr = np.zeros(10)
zero_arr[4] = 1  # 5th element (index 4)
print("\n6. Zero array with 5th element as 1:")
print(zero_arr)

# 7 Reverse array
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\n7. Reversed array:")
print(reversed_arr)

# 8 2D array with 1 on border and 0 inside
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n8. Border array (1 on border, 0 inside):")
print(border_array)

# 9 8x8 checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1  # Odd rows, even columns
checkerboard[::2, 1::2] = 1  # Even rows, odd columns
print("\n9. 8x8 Checkerboard pattern:")
print(checkerboard)
```

```python
# Array Operations:

import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

print("Original arrays:")
print(f"x = \n{x}")
print(f"y = \n{y}")
print(f"v = {v}")
print(f"w = {w}")

# 1 Add the two arrays
add_xy = x + y
print("\n1. x + y:")
print(add_xy)

# 2 Subtract the two arrays
sub_xy = x - y
print("\n2. x - y:")
print(sub_xy)

# 3 Multiply array with integer
mult_x = x * 3
print("\n3. x * 3:")
print(mult_x)

# 4 Square of each element
square_x = x ** 2
print("\n4. Square of x:")
print(square_x)

# 5 Dot products
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print(f"\n5. Dot products:")
print(f"v · w = {dot_vw}")
print(f"x · v = {dot_xv}")
print(f"x · y = \n{dot_xy}")

# 6 Concatenation
concat_row = np.concatenate((x, y), axis=0)
concat_col_vw = np.column_stack((v, w))
print("\n6. Concatenation:")
print(f"x and y along rows:\n{concat_row}")
print(f"v and w along columns:\n{concat_col_vw}")

# 7 Concatenate x and v (with error explanation)
try:
    concat_xv = np.concatenate((x, v))
    print(f"x and v concatenated:\n{concat_xv}")
except ValueError as e:
    print(f"\n7. Error concatenating x and v: {e}")
    print("Explanation: Arrays have different dimensions - x is 2D (2,2) and v is 1D (2,).")
    print("They must have the same number of dimensions for concatenation.")
```

```python
#Matrix Operations:
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
I = np.eye(2)  # Identity matrix

print("Given matrices:")
print(f"A = \n{A}")
print(f"B = \n{B}")

# 1 Prove A · A⁻¹ = I
A_inv = np.linalg.inv(A)
proof_1 = np.dot(A, A_inv)
print("\n1. Proof A · A⁻¹ = I:")
print(f"A · A⁻¹ = \n{proof_1}")
print(f"Identity matrix: \n{I}")
print(f"Are they equal? {np.allclose(proof_1, I)}")

# 2 Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. Proof AB ≠ BA:")
print(f"AB = \n{AB}")
print(f"BA = \n{BA}")
print(f"Are they different? {not np.array_equal(AB, BA)}")

# 3 Prove (AB)ᵀ = BᵀAᵀ
AB_transpose = AB.T
B_trans_A_trans = np.dot(B.T, A.T)
print("\n3. Proof (AB)ᵀ = BᵀAᵀ:")
print(f"(AB)ᵀ = \n{AB_transpose}")
print(f"BᵀAᵀ = \n{B_trans_A_trans}")
print(f"Are they equal? {np.array_equal(AB_transpose, B_trans_A_trans)}")

# Solving system of linear equations
print("\n4. Solving System of Linear Equations:")
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Matrix form: AX = B
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

print(f"Coefficient matrix A:\n{A_eq}")
print(f"Constants vector B: {B_eq}")

# Using inverse method: X = A⁻¹B
try:
    A_inv_eq = np.linalg.inv(A_eq)
    X_solution = np.dot(A_inv_eq, B_eq)
    print(f"\nSolution using inverse method (X = A⁻¹B):")
    print(f"x = {X_solution[0]:.2f}, y = {X_solution[1]:.2f}, z = {X_solution[2]:.2f}")

    # Verify solution
    verification = np.dot(A_eq, X_solution)
    print(f"Verification (A · X): {verification}")
    print(f"Matches B? {np.allclose(verification, B_eq)}")

except np.linalg.LinAlgError as e:
    print(f"Error: {e}")

# Using np.linalg.solve (more numerically stable)
print(f"\nSolution using np.linalg.solve:")
X_solve = np.linalg.solve(A_eq, B_eq)
print(f"x = {X_solve[0]:.2f}, y = {X_solve[1]:.2f}, z = {X_solve[2]:.2f}")
```
