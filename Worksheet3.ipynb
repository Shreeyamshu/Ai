Q5 – EDA Exercise 1

```python
# Q5.0 Fill missing Age
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
# Q5.1 Fare per year
df['fare_per_year']=df['Fare']/df['Age']
high_fare_age=df[df['fare_per_year']>5]
high_fare_age_srt=high_fare_age.sort_values('fare_per_year',ascending=False)
result=high_fare_age_srt[['Name','fare_per_year']]
result.head()
```

```python
# Q5.2 Fare per class for adult males
df['fare_per_class']=df['Fare']/df['Pclass']
adult_males=df[(df['Sex']=='male')&(df['Age']>=18)]
adult_males_srt=adult_males.sort_values('fare_per_class',ascending=False)
result2=adult_males_srt[['Name','Age','fare_per_class']]
result2.head()
```

Q6 – Group-By Analysis

```python
# Q6.1 Total fare per class
total_fare=df['Fare'].sum()
fare_class1=df[df['Pclass']==1]['Fare'].sum()
fare_class2=df[df['Pclass']==2]['Fare'].sum()
fare_class3=df[df['Pclass']==3]['Fare'].sum()
percent_revenue=[(x/total_fare)*100 for x in [fare_class1,fare_class2,fare_class3]]
percent_revenue
```

```python
# Q6.2 Age grouping percentage
def age_group(age):
    if age<18: return 'child'
    elif age<65: return 'adult'
    else: return 'senior'
df['age_group']=df['Age'].apply(age_group)
age_group_counts=df['age_group'].value_counts()
(age_group_counts/len(df))*100
```
